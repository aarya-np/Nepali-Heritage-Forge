```python
checkpoint_path = '/content/drive/MyDrive/translator/results/checkpoint-52602'
```

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("Swamitucats/M2M100_Sanskrit_English")
tokenizer = AutoTokenizer.from_pretrained("Swamitucats/M2M100_Sanskrit_English")
```

```python
from peft import PeftModel
lora_model = PeftModel.from_pretrained(model, checkpoint_path)
lora_tokenizer = AutoTokenizer.from_pretrained(checkpoint_path)
```

```python
sanskrit_text = ["कर्मण्येवाधिकारस्ते मा फलेषु कदाचन।",'सर्वं खल्विदं ब्रह्मा।','आत्मा वाङ्मयः।','विद्या ददाति विनयं विनयाद् याति पात्रताम्।']
inputs = lora_tokenizer(sanskrit_text, return_tensors="pt", padding = True, truncation = True)
outputs = lora_model.generate(**inputs, max_length=300)
english_translation = tokenizer.batch_decode(outputs, skip_special_tokens=True)
print(english_translation)

'''
Overall of each sentence : ["You have the right to perform your prescribed duties, but you are not entitled to the fruits of your actions.",
            'All this—everything you see—is indeed Brahman (the ultimate reality).',
            'The self (Atman) is made of words (language).',
            'Knowledge imparts humility, and through humility, one gains worthiness.']
'''
```
