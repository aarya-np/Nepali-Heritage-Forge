#BIBLE

```python
def create_csv(path1, path2):
  english_lines = open(path1).readlines()
  sanskrit_lines = open(path2).readlines()
  data = {'src': sanskrit_lines, 'tgt':english_lines}
  return data
```

```python
data = create_csv('/content/bible.en','/content/bible.sa')
df = pd.DataFrame(data)
df.to_csv('/content/drive/MyDrive/Colab Notebooks/Hugging Face Course/translator/datasets/bible.csv', index = False)
```
